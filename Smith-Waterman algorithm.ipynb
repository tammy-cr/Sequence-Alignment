```python
def smith_waterman(seq1, seq2, match_score=2, mismatch_score=-1, gap_penalty=-1):
    # Initialize the scoring matrix
    rows, cols = len(seq1) + 1, len(seq2) + 1
    score_matrix = [[0 for _ in range(cols)] for _ in range(rows)]

    # Initialize variables to keep track of the maximum score and its position
    max_score = 0
    max_i, max_j = 0, 0

    # Fill in the scoring matrix
    for i in range(1, rows):
        for j in range(1, cols):
            if seq1[i - 1] == seq2[j - 1]:
                match = score_matrix[i - 1][j - 1] + match_score
            else:
                match = score_matrix[i - 1][j - 1] + mismatch_score
            delete = score_matrix[i - 1][j] + gap_penalty
            insert = score_matrix[i][j - 1] + gap_penalty

            # Calculate the score for the current position
            score_matrix[i][j] = max(0, match, delete, insert)

            # Update the maximum score and its position
            if score_matrix[i][j] > max_score:
                max_score = score_matrix[i][j]
                max_i, max_j = i, j

    # Traceback to find the alignment
    alignment1, alignment2 = [], []

    i, j = max_i, max_j

    while i > 0 and j > 0 and score_matrix[i][j] > 0:
        if seq1[i - 1] == seq2[j - 1]:
            alignment1.append(seq1[i - 1])
            alignment2.append(seq2[j - 1])
            i -= 1
            j -= 1
        elif score_matrix[i][j] == score_matrix[i - 1][j - 1] + mismatch_score:
            alignment1.append(seq1[i - 1])
            alignment2.append(seq2[j - 1])
            i -= 1
            j -= 1
        elif score_matrix[i][j] == score_matrix[i - 1][j] + gap_penalty:
            alignment1.append(seq1[i - 1])
            alignment2.append('-')
            i -= 1
        else:
            alignment1.append('-')
            alignment2.append(seq2[j - 1])
            j -= 1

    # Reverse the alignment sequences
    alignment1 = ''.join(reversed(alignment1))
    alignment2 = ''.join(reversed(alignment2))

    return alignment1, alignment2, max_score

# Example usage:
seq1 = "AGTACGCA"
seq2 = "TATGC"
alignment1, alignment2, score = smith_waterman(seq1, seq2)
print("Alignment 1:", alignment1)
print("Alignment 2:", alignment2)
print("Alignment Score:", score)
```
